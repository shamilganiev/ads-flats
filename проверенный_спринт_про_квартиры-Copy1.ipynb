#### Обработка `ceiling_height`

> Потолки выше 4 метров и ниже 2.4 примем как выбросы (СНиП)

```python
max_c = 4
min_c = 2.4
is_abnormal_ceiling = (df['ceiling_height'] > max_c) | (df['ceiling_height'] < min_c)
abnormal_ceilings = df.loc[is_abnormal_ceiling]
display(abnormal_ceilings.head())
abnormal_ceilings_count = len(abnormal_ceilings)
print(f'Колоичество ОН с потолками (< {min_c}) или (> {max_c}) метров: {abnormal_ceilings_count}\n' + \
      f'Процент от общего количества: {abnormal_ceilings_count / len(df["ceiling_height"]) * 100}') 
```

```python
print(f'Процент аномальных ОН около {abnormal_ceilings_count / len(df["ceiling_height"]) * 100}%, так что можно спокойно удалить эти строки, на общую картину это не повлияет')
```

```python
df.drop(df[is_abnormal_ceiling].index, errors='ignore', inplace=True)
```

## Расчёты и добавление результатов в таблицу

```python
def price_per_sqmeter(row):
    return row['last_price']/row['total_area']
# цена за квадратный метр
```

```python
df['price_per_meter'] = df.apply(price_per_sqmeter, axis=1)
```

```python
df.head()
```

Создаем столбы: неделя продажи, месяц продажи, год продажи

```python
df['weekday'] = df['day_exposition'].dt.dayofweek + 1  #     (1 - Пн, 2 - Вт, ... , 7 - Вс)
df['month'] = df['day_exposition'].dt.month
df['year'] = df['day_exposition'].dt.year
```

```python
df.head()
```

Категория этажности

```python
def floor_cat(row):
    if row['floor'] == 1:
        return 'первый'
    elif row['floors_total'] == row['floor']:
        return 'последний'
    else:
        return 'другой'
```
